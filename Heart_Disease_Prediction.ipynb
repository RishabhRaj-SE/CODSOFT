Importing the dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and PreProcessing


```python
heart_data=pd.read_csv("/content/heart_disease_data.csv")
```

```python
heart_data.head()
```

```python
heart_data.describe()
```

```python
heart_data.shape
```

```python
heart_data.info()
```

```python
heart_data.isnull().sum()
```

Check the distribution of target variable


```python
heart_data["target"].value_counts()
```

here the distribution count is ok so no need for undersampling or overSampling

```python
X=heart_data.drop(columns="target",axis=1)
Y=heart_data["target"]
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=20)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

Model Training

```python
Logistic_model=LogisticRegression()
Logistic_model.fit(X_train,Y_train)
```

```python
X_train_prediction=Logistic_model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction=Logistic_model.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuarcy)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model=RandomForestClassifier(n_estimators=100,random_state=42)
rf_model.fit(X_train,Y_train)
```

```python
X_train_prediction=rf_model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(training_data_accuracy)
```

```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'   # prevents warning
)
```

```python
xgb_model.fit(X_train,Y_train)
```

```python
X_train_prediction=xgb_model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print(training_data_accuracy)
```

```python
X_test_prediction=xgb_model.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print(test_data_accuarcy)
```

```python
X_test_prediction=rf_model.predict(X_test)
test_data_accuarcy=accuracy_score(X_test_prediction,Y_test)
print(test_data_accuarcy)
```
